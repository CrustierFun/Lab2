```python
###Задание №3
import time
import math
import random

m = input('m = ')
n = input('n = ')
m = int(m)
n = int(n)
if m <= 0:
    m = 50
if n <= 0:
    n = 50

min_limit = input('min_limit = ')
max_limit = input('max_limit = ')
min_limit = int(min_limit)
max_limit = int(max_limit)
if min_limit == 0:
    min_limit = -250
if max_limit == 0:
    max_limit = 1000+15
if min_limit > max_limit:
    min_limit = -250
    max_limit = 1000 + 6

Matrix = [[random.randint(min_limit, max_limit) for j in range(n)] for i in range(m)]
print('Matrix:')
a=b=c=d=e=f=g=Matrix
for i in range(m):
    print(Matrix[i])
    
# Сортировка выбором
def selection_sort(array):
    for i in range(len(array)):
        low_index = i
        for j in range(i + 1, len(array)):
            if array[j] < array[low_index]:
                low_index = j
        array[i], array[low_index] = array[low_index], array[i]
        
start_time = time.time()        
print('')
print('sort_viborom')
for i in range(m):
    selection_sort(a[i])
    print(a[i])
print('')
print(f"{(time.time() - start_time)} секунд")    
print('')

# Сортировка вставкой
def insertion_Sort(array):
    for i in range(1, len(array)):
        key = array[i]
        j = i-1
        while array[j] > key and j >= 0:
            array[j+1] = array[j]
            j -= 1
        array[j+1] = key
    return array 

start_time = time.time()
print('sort_vstavkoy:')
for i in range(m):
    insertionSort(b[i])
    print(b[i])
print('')
print(f"{(time.time() - start_time)} секунд")    
print('')

# Сортировка пузырьком
def bubble_Sort(array):
    swapped = False
    for i in range(len(array)-1,0,-1):
        for j in range(i):
            if array[j]>array[j+1]:
                array[j], array[j+1] = array[j+1], array[j]
                swapped= True
        if swapped:
            swapped=False
        else:
            break
    return array
start_time = time.time()
print('sort_bubble:')
for i in range(m):
    bubble_Sort(c[i])
    print(c[i])
print('')
print(f"{(time.time() - start_time)} секунд")    
print('')

# Сортировка Шелла
def shell_Sort(array):
    n = len(array)
    k = int(math.log2(n))
    interval = 2**k -1
    while interval > 0:
        for i in range(interval, n):
            temp = array[i]
            j = i
            while j >= interval and array[j - interval] > temp:
                array[j] = array[j - interval]
                j -= interval
            array[j] = temp
        k -= 1
        interval = 2**k -1
    return array
start_time = time.time()
print('sort_Shella:')
for i in range(m):
    shell_Sort(d[i])
    print(d[i])
print('')
print(f"{(time.time() - start_time)} секунд")    
print('')

# Турнирная сортировка
def merge_Sort(nums):
    if len(nums)==1:
        return nums
    mid = (len(nums)-1) // 2
    lst1 = merge_Sort(nums[:mid+1])
    lst2 = merge_Sort(nums[mid+1:])
    result = merge(lst1, lst2)
    return result
def merge(lst1, lst2):
    lst = []
    i = 0
    j = 0
    while(i<=len(lst1)-1 and j<=len(lst2)-1):
        if lst1[i]<lst2[j]:
            lst.append(lst1[i])
            i+=1
        else:
            lst.append(lst2[j])
            j+=1
    if i>len(lst1)-1:
        while(j<=len(lst2)-1):
            lst.append(lst2[j])
            j+=1
    else:
        while(i<=len(lst1)-1):
            lst.append(lst1[i])
            i+=1
    return lst
start_time = time.time()
print('sort_turnir:')
for i in range(m):
    merge_Sort(e[i])
    print(e[i])
print(' ')
print(f"{(time.time() - start_time)} секунд")    
print('')

# Быстрая сортировка
def quick_Sort(Matrix):
    array = Matrix.copy()
    if len(array) < 2:
        return array
    low = []
    same = []
    high = []
    ignat = array[random.randint(0, len(array) - 1)]
    for item in array:
        if item < ignat:
            low.append(item)
        elif item == ignat:
            same.append(item)
        elif item > ignat:
            high.append(item)
    return quick_Sort(low) + same + quick_Sort(high)
    
start_time = time.time()
print(' ')
print('sort_fast: ')
for i in range(m):
    print(quick_Sort(Matrix[i]))
print(' ')
print(f"{(time.time() - start_time)} секунд")
print('')

# Пирамидная сортировка
def heapify_Sort(array, n, i):
    largest = i
    l = 2 * i + 1
    r = 2 * i + 2

    if l < n and array[i] < array[l]:
        largest = l
    if r < n and array[largest] < array[r]:
        largest = r

    if largest != i:
        array[i], array[largest] = array[largest], array[i]
        heapify_Sort(array, n, largest)
def heapSort(array):
    n = len(array)
    for i in range(n // 2, -1, -1):
        heapify_Sort(array, n, i)
    for i in range(n - 1, 0, -1):
        array[i], array[0] = array[0], array[i]
        heapify_Sort(array, i, 0)
    return array

start_time = time.time()
print(' ')
print('sort_pyramid:')
for i in range(m):
    heapSort(g[i])
    print(g[i])
print(f"{(time.time() - start_time)} секунд")
print(' ')

start_time = time.time()
for i in range(m):
    sorted(f[i])
    print(f[i])
print(f"{(time.time() - start_time)} секунд")
```
